REVERSING A STRING

```python
str  = input("Enter string : ")
rev = str[::-1]
print(rev)
```

CHECKING PALINDROME

```python
i = input("Enter string : ")
if i == i[::-1]:
    print(i," is an palindrome")
else:
    print(i," is not an palindrome")
```

FINDING LARGEST AND SMALLEST NUMBER IN LIST

```python
list = [22,53,76,98,13]
la = max(list)
print(la)
sl = min(list)
print(sl)
```
